```python
from tensorflow.keras.applications import VGG16
```

```python
model=VGG16(weights='imagenet')
```

```python
model.summary()
```

```python
#load image
import matplotlib.pyplot as plt
import matplotlib.image as img
image=img.imread('/content/drive/MyDrive/PGP Datasets/Doggy1.jpeg')
```

```python
plt.imshow(image)
```

```python
#preprocessing the image
imgpath=('/content/drive/MyDrive/PGP Datasets/ObamaBo.jpg')
print('original image shape',img.imread(imgpath).shape)
```

```python
from tensorflow.keras.preprocessing import image as image_utils
from tensorflow.keras.applications.vgg16 import preprocess_input
```

```python
#convert image into target size of 224x224
image=image_utils.load_img(imgpath,target_size=(224,224))
#convert to array
image=image_utils.img_to_array(image)
#reshape with colour channel
image=image.reshape(1,224,224,3)
#preprocess the input
image=preprocess_input(image)
print(image.shape)
```

```python
#predict the model
predictions=model.predict(image)
```

```python
from tensorflow.keras.applications.vgg16 import decode_predictions
```

```python
print('predicted',decode_predictions(predictions,top=3))
```
